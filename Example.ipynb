```python
import matplotlib.pyplot as plt
from sklearn import datasets
import numpy as np
from ClusterAccuracy.ClusteringAccuracy import clustering_accuracy_match, clustering_accuracy_pair, clustering_accuracy_match_min
```

```python
from DensityBasedClusteirngAlgorithms import HaloBasedClustering 
```

```python
blobs = datasets.make_blobs(n_samples=[300,500,500],
                            centers=[[-0.7,5.0],[5.5,7],[-0.0,8]],
                             cluster_std=[1.1, 2.5,0.4 ],
                             random_state=27)
X1, y1 = blobs
plt.figure(figsize=(2.5,2),dpi=200)
plt.scatter(X1[:, 0], X1[:, 1],s=0.1,c='black',marker='o',alpha=0.8)
plt.title("G3",fontsize=10,fontfamily='Times New Roman')
plt.xticks(())
plt.yticks(())
plt.axis('off')
```

```python
X=X1
y=y1

HBC=HaloBasedClustering.HaloBasedClustering(n_neighbors=40,
        density_threshold_pct=40,tol=1
        )
HBC_y=HBC.fit_predeict(X1)
```

```python
clustering_accuracy_match_min(HBC_y,y1),clustering_accuracy_match_min(HBC_y[HBC_y!=-1],y1[HBC_y!=-1])
```
